```python
import random
import math

selected = input('ODD OR EVEN: ').upper()
cpu_input = math.ceil(random.random() * 6)
player_input = int(input('DO TOSS: '))

print('TOSS: YOU: ' + str(player_input))
print('TOSS: COMPUTER: ' + str(cpu_input))

status = ""
if selected == 'EVEN':
    if (cpu_input + int(player_input)) % 2 == 0:
        status = input("BATTING OR BOWLING? ").upper()
    else:
        choice = random.random()
        if choice < 0.5:
            print('COMPUTER WON THE TOSS AND CHOSE TO BAT')
            status = 'BOWLING'
        else:
            print('COMPUTER WON THE TOSS AND CHOSE TO BOWL')
            status = 'BATTING'
elif selected == 'ODD':
    if (cpu_input + int(player_input)) % 2 == 1:
        status = input('BATTING OR BOWLING? ').upper()
    else:
        choice = random.random()
        if choice < 0.5:
            print('COMPUTER WON THE TOSS AND CHOSE TO BAT')
            status = 'BOWLING'
        else:
            print('COMPUTER WON THE TOSS AND CHOSE TO BOWL')
            status = 'BATTING'

score = 0
target = 1
innings = 1

while True:
    player_input = int(input('YOU: ' + status + ' INNINGS ' + str(innings) + ': '))
    cpu_input = math.ceil(random.random() * 6)
   
    print('YOU: ' + str(player_input))
    print('COM: ' + str(cpu_input))

    if player_input == cpu_input:
        if status == 'BATTING' and innings == 1:
            print('OUTTT!!')
            target = score + 1
            print('TO DEFEND: ' + str(target))
            status = 'BOWLING'
            innings = 2
        elif status == 'BOWLING' and innings == 1:
            print('OUTTT!!!')
            target = score + 1
            print('TO GET: ' + str(target))  
            status = 'BATTING'
            innings = 2
        elif status == 'BATTING' and innings == 2:
            print('OUTTT...YOU LOST THE MATCH')
            break
        elif status == 'BOWLING' and innings == 2:
            print('OUT...YOU WON THE GAME')
            break
    else:
        if status == 'BATTING' and innings == 1:
            score += player_input
            print('SCORE: ' + str(score))
        elif status == 'BATTING' and innings == 2:
            if player_input >= target:
                print('YOU WON THE MATCH!')
                break
            else:
                target -= player_input
                print('TO GET: ' + str(target))
        elif status == 'BOWLING' and innings == 1:
            score += cpu_input
            print('SCORE: ' + str(score))
        elif status == 'BOWLING' and innings == 2:
            if cpu_input >= target:
                print("YOU LOST THE MATCH")
                break
            else:
                target -= cpu_input
                print('TO DEFEND: ' + str(target))  

```
